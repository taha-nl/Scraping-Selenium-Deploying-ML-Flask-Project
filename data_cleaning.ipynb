###  salary parsing

```python
(df["Salary Estimate"]=="-1").sum()
```

```python
df=df[df["Salary Estimate"]!="-1"]
```

```python
df["Salary Estimate"].sample(10)
```

```python
df["Salary Estimate"].sample(30)
```

```python
df["hourly"]=df["Salary Estimate"].apply(lambda x:1 if "per hour" in x.lower() else 0)
```

```python
df["employer_provided"]=df["Salary Estimate"].apply(lambda x:1 if "employer provided salary" in x.lower() else 0)
```

```python
df.sample(10)
```

```python
salary=df["Salary Estimate"].str.split("(").apply(lambda x:x[0].replace("$","").replace("K",""))
```

```python
salary=salary.apply(lambda x:x.split(":")[1] if len(x.split(":")) >1 else x)
```

```python
salary=salary.apply(lambda x:x.lower().replace("per hour",""))
```

```python
df["range-salary"]=salary
```

```python
df["min_salary"]=salary.apply(lambda x:x.split("-")[0])
df["max_salary"]=salary.apply(lambda x:x.split("-")[1])
```

```python
df["avg_salary"]=(df["min_salary"].astype('int')+df["max_salary"].astype('int'))/2
```

#### state field

```python
df["job_state"]=df["Location"].apply(lambda x:x.split(",")[1])
```

```python
df.job_state.value_counts()
```

```python
df["same_state"]=df.apply(lambda x: 1 if  x["Location"]==x["Headquarters"] else 0,axis=1)
```

#### age of the company

```python
df["age"]=df["Founded"].apply(lambda x:x if x<1 else 2023-x)
```

```python
df["python_skill"]=df["Job Description"].apply(lambda x:1 if "python" in x.lower() else 0)
df["rstudio_skill"]=df["Job Description"].apply(lambda x:1 if "r studio" in x.lower() else 0)
df["spark_skill"]=df["Job Description"].apply(lambda x:1 if "spark" in x.lower() else 0)
df["aws_skill"]=df["Job Description"].apply(lambda x:1 if "aws" in x.lower() else 0)
df["excel_skill"]=df["Job Description"].apply(lambda x:1 if "excel" in x.lower() else 0)
```

### parsing size

```python
df["Size"].sample(10)
```

```python
import re
```

```python
df["size_number"]=df["Size"].apply(lambda x: re.search(r"\d+\sto\s\d+|(\d+)\+",x).group() if re.search(r"\d+\sto\s\d+|(\d+)\+",x) else x)
```

```python
df["size_number"]=df["size_number"].apply(lambda x:x.replace(r" to ","-"))
```
